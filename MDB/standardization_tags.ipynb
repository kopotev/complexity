```python
import os
import glob
import re
import matplotlib.pyplot as plt

folder_path = '/content/Students_texts'
file_list = glob.glob(os.path.join(folder_path, "*.txt"))

counts = []
for fname in file_list:
  with open(fname,'r', encoding='utf-8') as f:
    text = f.read()
    count = len(re.findall(r'<.*?>', text))
    counts.append(count)

plt.bar(range(len(file_list)), counts, tick_label=file_list)
plt.xlabel('Файлы')
plt.ylabel('standardization tags')
plt.title('Number of tags in each file')
plt.xticks(rotation=90)
#plt.tight_layout()
plt.show()
```
