```python
import os
import glob
import matplotlib.pyplot as plt
import numpy as np

# Путь к папке с файлами txt
folder_path = "/content"

# Получаем список всех файлов txt в папке
file_list = glob.glob(os.path.join(folder_path, "*.txt"))

# Список для хранения средних длин токенов
avg_token_lengths = []

# Обрабатываем каждый файл
for file_path in file_list:
    with open(file_path, 'r') as file:
        text = file.read()
        tokens = text.split()  # Разделение текста на токены
        token_lengths = [len(token) for token in tokens]  # Длины токенов в символах
        avg_token_length = np.mean(token_lengths)  # Средняя длина токена
        avg_token_lengths.append(avg_token_length)

# Создание столбчатой диаграммы
plt.bar(range(len(file_list)), avg_token_lengths)
plt.xlabel('Файлы')
plt.ylabel('Средняя длина токена')
plt.title('Средняя длина токена в каждом файле')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
